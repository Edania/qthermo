```python
import numpy as np
import thermo_funcs as tf
import matplotlib.pyplot as plt

n_f = 5
E = np.linspace(-10,5)
mu = np.random.uniform(-5,0,n_f).reshape(-1,1)#np.linspace(-5,0,n_f).reshape(-1,1)
T = np.random.uniform(0.1,10,n_f).reshape(-1,1)#np.linspace(0.1,10,n_f).reshape(-1,1)
c = np.random.uniform(0,1,n_f)
c = c/np.sum(c)
c_con = np.random.uniform(0.5,1,n_f)
c_con = c_con/np.sum(c_con)

f = tf.fermi_dist(E,mu,T)
print(((c*f.T).T).shape)
f_tot = np.sum((c*f.T).T, axis = 0)
f_con = 1/(1+np.exp(-(E-mu)/T))
f_con_tot = np.sum((c_con*f_con.T).T, axis = 0)
plt.plot(E,f_tot, label = "Fermi")
plt.plot(E,f_con_tot, label = "Fermi con")
plt.plot(E, f_con_tot + f_tot)
plt.show()
```
